```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
from keras.models import Sequential
from keras.layers import Dense

dt = pd.read_excel('Real estate valuation data set.xlsx')
data = dt.loc[:,"X1 transaction date":"Y house price of unit area"]

'''---------------------------input data-----------------------------'''

train_x = data.sample(frac=0.8,random_state=0)
test_x = data.drop(train_x.index)

#y_data = p1["Raw Annual Return"]
train_y = train_x.pop('Y house price of unit area')
test_y = test_x.pop('Y house price of unit area')
#y_norm_data = p1.loc[:,"Norm Annual Return":"Norm Rel. Win Rate"]

#Model, Cost, Train
model = Sequential()

#Dense(out, in)
model.add(Dense(1, input_dim = 6))
model.compile(loss=tf.keras.losses.MeanSquaredLogarithmicError(), optimizer='sgd', metrics=['accuracy'])
model.fit(train_x, train_y, epochs=6)
model.summary()

#Inference
# print(model.get_weights())
print(model.predict(test_x))
print(model.evaluate(test_x,test_y))
```
